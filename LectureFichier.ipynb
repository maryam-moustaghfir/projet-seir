# Lecture de quelques fichiers csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Rpertoire o se trouvent les fichiers CSV de chaque rplication
directory = r'C:\Users\Aymen\Desktop\projet\SEIR' 

# Liste des fichiers CSV dans le rpertoire
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Nombre total d'individu 
reference_value = 20000    # Vous pouvez changer cette valeur en fonction de vos besoins

# Boucle pour traiter chaque fichier CSV de rplication
for csv_file in csv_files:
    # Lire les donnes depuis le fichier CSV
    data = pd.read_csv(os.path.join(directory, csv_file))

    # Extraire les donnes ncessaires
    iteration_values = data['Iteration']
    countS_values = data['CountS']
    countE_values = data['CountE']
    countI_values = data['CountI']
    countR_values = data['CountR']

    # Tracer le graphique
    plt.figure(figsize=(9, 4))
    plt.plot(iteration_values, countS_values / reference_value, label='susceptible', color='blue')
    plt.plot(iteration_values, countE_values / reference_value, label='exposed', color='orange')
    plt.plot(iteration_values, countI_values / reference_value, label='infected', color='green')
    plt.plot(iteration_values, countR_values / reference_value, label='recoverd', color='red')

    # Ajouter des tiquettes au graphique
    plt.title(csv_file)
    plt.xlabel('Time in days')
    plt.ylabel('Ratio of agents in each state')
    plt.grid(True)
    plt.legend()
    plt.show()
```
