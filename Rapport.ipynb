Vous trouverez ce dernier ci dessous. 


```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Répertoire où se trouvent les fichiers CSV de chaque réplication
directory = r'C:\Users\Aymen\Desktop\projet\SEIR'

# Liste des fichiers CSV dans le répertoire
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# Initialisation des listes pour stocker les données de toutes les réplications
all_countS_values = []
all_countE_values = []
all_countI_values = []
all_countR_values = []

# Nombre total d'individus
reference_value = 20000    # Vous pouvez changer cette valeur en fonction de vos besoins

# Boucle pour traiter chaque fichier CSV de réplication
for csv_file in csv_files:
    # Lire les données depuis le fichier CSV
    data = pd.read_csv(os.path.join(directory, csv_file))

    # Extraire les données nécessaires
    countS_values = data['CountS']
    countE_values = data['CountE']
    countI_values = data['CountI']
    countR_values = data['CountR']

    # Ajouter les données de cette réplication aux listes globales
    all_countS_values.append(countS_values)
    all_countE_values.append(countE_values)
    all_countI_values.append(countI_values)
    all_countR_values.append(countR_values)

# Calcul de la moyenne des valeurs pour chaque état
mean_countS = sum(all_countS_values) / len(all_countS_values)
mean_countE = sum(all_countE_values) / len(all_countE_values)
mean_countI = sum(all_countI_values) / len(all_countI_values)
mean_countR = sum(all_countR_values) / len(all_countR_values)

# Tracer le graphique moyen
plt.figure(figsize=(9, 4))
plt.plot(mean_countS, label='susceptible', color='blue')
plt.plot(mean_countE, label='exposed', color='orange')
plt.plot(mean_countI, label='infected', color='green')
plt.plot(mean_countR, label='recovered', color='red')

# Ajouter des étiquettes au graphique
plt.title('Moyenne de toutes les réplications')
plt.xlabel('Time in days')
plt.ylabel('Ratio moyen des agents dans chaque état')
plt.legend()
plt.grid(True)
plt.show()
```
